## Where Python is Used?

#### Python has a rich ecosystem of libraries and frameworks that cater to a wide range of domains and purposes. Here are some of the most commonly used libraries in Python across different fields:

#### * NumPy: Essential for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.

#### * Pandas: A powerful library for data manipulation and analysis. It provides data structures like DataFrames and Series, which are highly efficient for handling structured data.

#### * Matplotlib: A plotting library for creating static, animated, and interactive visualizations in Python. It enables the creation of a wide variety of charts and plots.

#### * Scikit-learn: A leading library for machine learning in Python. It provides simple and efficient tools for data mining and data analysis, implementing a wide variety of machine learning algorithms.

#### * TensorFlow / PyTorch: These are deep learning libraries used for building and training neural network models. TensorFlow is developed by Google and is widely used for production-level tasks, while PyTorch is favored for its flexibility and ease of use in research.

#### * Keras: A high-level neural networks API, written in Python and capable of running on top of TensorFlow, CNTK, or Theano. Keras simplifies the process of building deep learning models.

#### * SciPy: A library used for scientific and technical computing. It builds on NumPy and provides a large number of functions that operate on NumPy arrays and are useful for different types of scientific and engineering applications.

#### * NLTK (Natural Language Toolkit): A library for working with human language data. It provides easy-to-use interfaces to over 50 corpora and lexical resources such as WordNet, along with a suite of text processing libraries for classification, tokenization, stemming, tagging, parsing, and more.

#### * Beautiful Soup: A library used for web scraping and parsing HTML and XML documents. It creates parse trees from page source code that can be used to extract data easily.

#### * Django / Flask: These are web frameworks for Python. Django is a high-level web framework that encourages rapid development and clean, pragmatic design. Flask is a lightweight framework that is easy to use and extend.

#### * Requests: An elegant and simple HTTP library for Python, used for making HTTP requests. It simplifies the process of sending HTTP requests and handling responses.

#### * Pygame: A set of Python modules designed for writing video games. It includes computer graphics and sound libraries and is commonly used for educational purposes and prototyping.

## Rules for variable names 

#### - names can not start with a number
#### - names can not contain spaces, use _ intead
#### - names can not contain any of these symbols: '",<>/?|\!@#%^&*~-+
#### - it's considered best practice (PEP8) that names are lowercase with underscores
#### - avoid using Python built-in keywords like list and str
#### - Avoid using the single characters l (lowercase letter el), O (uppercase letter oh) and I (uppercase letter eye)  as they can be confused with 1 and 0
