```python
import pandas as pd
import numpy as np

# Step 1: Create a sample dataset mimicking "student_scores.csv"
data = {
    "Student_ID": [1, 2, 3, 4, 5, 6],
    "Name": ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    "Math_Score": [78, 85, 92, 88, 79, 76],
    "Science_Score": [85, 82, 91, 79, 95, 84],
    "English_Score": [92, 88, 85, 78, 91, 77]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Save as "student_scores.csv" (optional)
df.to_csv('student_scores.csv', index=False)

# Step 1(a): Create a NumPy array for the first 5 students' Math scores
math_scores_array = np.array(df['Math_Score'][:5])

# Step 1(b): Perform operations on the NumPy array
# i) Find the mean of the Math scores
mean_math_score = np.mean(math_scores_array)

# ii) Find the maximum and minimum Math scores
max_math_score = np.max(math_scores_array)
min_math_score = np.min(math_scores_array)

# Output results
print("Math Scores Array for the first 5 students:", math_scores_array)
print("Mean of the Math scores:", mean_math_score)
print("Maximum Math score:", max_math_score)
print("Minimum Math score:", min_math_score)

# Step 2: Load the dataset using Pandas
# a. Display the first 10 rows of the dataset
print("\nFirst 10 rows of the dataset:")
print(df.head(10))

# b. Sort the data by Science_Score in descending order
sorted_by_science = df.sort_values(by='Science_Score', ascending=False)
print("\nDataset sorted by Science_Score in descending order:")
print(sorted_by_science)

# c. Filter and display the students who scored above 80 in English
students_above_80_english = df[df['English_Score'] > 80]
print("\nStudents who scored above 80 in English:")
print(students_above_80_english)

# d. Add a new column called "Total_Score"
df['Total_Score'] = df['Math_Score'] + df['Science_Score'] + df['English_Score']
print("\nUpdated dataset with 'Total_Score' column:")
highest_score = df['Total_Score']

max_highest_score = np.max(highest_score)
print(df.head())
# ii) Find the highest and minimum Math scores
print(f'Highest Score : {max_highest_score}')


```
